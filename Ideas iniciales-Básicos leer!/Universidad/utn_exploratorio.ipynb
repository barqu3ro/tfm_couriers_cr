```python
import matplotlib.pyplot as plt

# Group the data by distrito and count the number of occurrences
distrito_counts = data['Distrito'].value_counts()

# Plot the bar graph
plt.figure(figsize=(12, 6))
distrito_counts.plot(kind='bar')
plt.title('Number of Students per Distrito')
plt.xlabel('Distrito')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
