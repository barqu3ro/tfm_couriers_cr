# Entendimiento del negocio

## Determinación de los objetivos de negocio

### Antecedentes

Existen diversas aserciones para definir Comercio Electrónico (CE). Entre ellas, la Organización para la Cooperación y el Desarrollo Económico lo define como el proceso de compra, venta o intercambio de bienes, servicios e información a través de redes de comunicación, comunmente Internet. La clasificación más básica del CE se hace en base al tipo de entes que se relacionan: empresas (businesses, B), consumers (consummers, C) y entes públicos (governments, G). De esta forma, una empresa de CE convencional suele ser B2B si vende a otras empresas, B2G si su relación comercial es con administraciones o B2C si vende consumidores finales.

En nuestro caso, la empresa “Beauty eSheep” se puede considerar un CE de tipo B2C. Su producto estrella es una crema hidratante unisex con mucho éxito entre su clientela. A partir de este producto inicial, la empresa ha ido ofreciendo un catálogo de productos tanto de tanto de belleza como de bienestar y salud.

Hace tiempo la empresa instauró una estrategia relacional, centrada en el cliente, de tal manera que ha ido recabando diversos datos sobre los mismos incluidas las distintas compras que han realizado.

### Objetivos de negocio

Basándose en los datos recopilados para cada cliente, la empresa quiere realizar una campaña para impulsar la venta de tensiómetros digitales. La empresa tiene acceso a un stock muy flexible en fechas de envío de estos productos y a muy buen precio, por lo que se espera una buena rentabilidad en su venta.

Por tanto, en este proyecto hay que identificar el público objetivo susceptible de comprar dicho producto para ofrecerselo a través de la plataforma de CE de la compañía, SMS y/o webmail durante el periodo que dura la campaña que es de una semana. Una vez identificado el cliente apropiado para ofrecerle el producto, se le va a regalar una suscripción a una aplicación de salud durante un año, esta aplicación incentiva el uso de los tensiómetros digitales y, por tanto, se espera que sea un incentivo para la compra del mismo. Este regalo es incondicional (no se le exige que compre el tensiómetro en cuestión). La suscripción tiene un coste para la compañía de 5 euros.

### Criterios de éxito

Se ha especificado en el fichero “data/raw/PAR_NEGOCIO.csv” diversos parámetros de usuario para este proyecto. En concreto se han especificado los siguientes indicadores o KPIs:

- NUEVOS CLIENTES ESTIMADOS CAMPANA: Clientes que se han estimado comprarán en nuestro empresa de forma eletrónica durante el periodo de campaña. Todos esos clientes serán susceptibles de ofrecerles el producto PRO13 si se identifican como idóneos para dicho producto. La empresa tiene una fuerte estrategia relacional de tal manera que no quieren ofrecer productos a clientes que entiendan no son adecuados.
- BENEFICIO NETO OBJETIVO: Importe en euros del beneficio neto que espera obtener tras la realización de la campaña.


## Evaluación de la situación

Como se ha comentado, se trata de una empresa de CE que además basa la toma de decisiones en los datos. Para conseguir llevar a cabo dicho paradigma la empresa cuenta con la arquitectura conceptual que se muestra a continuación:



Arquitectura conceptual de la empresa

Los componentes de esta arquitectura conceptual son los siguientes (Galán et al., 2022):

- Soluciones de gestión de datos para el análisis empresarial. Los datos son la materia prima esencial para la toma de decisiones basada en datos. A partir de ellos, los algoritmos de Machine Learning (ML) podrán encontrar el conocimiento deseado en forma de patrones. De esta manera, debemos tener acceso a los datos tanto internos como externos que se consideren importantes para la toma de decisiones. El formato de estos datos puede ser estructurado (típicamente tablas de bases de datos relacionales que se manipulan con SQL), semiestructurado (con estructura pero sin tablas, incluyendo las bases de datos NoSQL) o no estructurado (sin un formato definido, como lenguaje natural, imágenes, vídeo, etc.). Una organización convencional se caracteriza principalmente por los datos internos estructurados de sus sistemas operativos: ERP (Enterprise Resources Planing, Planificación de Recursos Empresariales) y CRM (Customer Relationship Management, gestión relacional de clientes). Antes de su análisis, estos datos deben ser almacenados. Los algoritmos de ML se clasifican en tres categorías principales: aprendizaje supervisado, aprendizaje no supervisado y aprendizaje por refuerzo. El ML supervisado utiliza datos etiquetados en una determinada clase objetivo, que debe ser aprendida por el algoritmo. Si esta variable objetivo es continua estaríamos en un caso de regresión, al igual que si es discreta en clasificación. El ML no supervisado trabaja con datos no etiquetados obteniendo agrupaciones de los datos, reglas de asociación, reducción de la dimensionalidad, etc. El ML de refuerzo considera el problema de un agente computacional que aprende a tomar decisiones por un método de prueba y error.

```python
# apertura de la base de datos y generar cursor
conexion=sqlite3.connect("../data/final/dw_sf_tr.db")
cursor=conexion.cursor()
```

```python
# Productos vendidos a cada cuenta
sql="""
    SELECT IDE_CUENTA, 
           REPLACE(COD_PRODUC||'_'||DES_PRODUC,' ','') DES_PRODUC, 
           NUM_PRECTOT_LIN_OPORTU
    FROM
        (SELECT IDE_CUENTA, IDE_PRODUC, SUM(NUM_PRECTOT_LIN_OPORTU) NUM_PRECTOT_LIN_OPORTU
        FROM FAC_LIN_OPORTU F
        INNER JOIN DIM_OPORTU O ON (F.IDE_OPORTU=O.IDE_OPORTU)
        GROUP BY IDE_CUENTA, IDE_PRODUC) VE
    INNER JOIN DIM_PRODUC PR
    ON VE.IDE_PRODUC=PR.IDE_PRODUC
"""
df_venta=pd.read_sql_query(sql, conexion)
df_venta.head(2)
```

```python
# Datos de los clientes
sql="""
    SELECT CU.IDE_CUENTA, 
           NUM_EDAD_CUENTA, 
           NUM_TAMANO_FAMILIAR_CUENTA, 
           NUM_ANOS_EXPERIENCIA_CUENTA, 
           NUM_INGRESOS_ANO_CUENTA, 
           DES_NIVEDU_CONTAC
    FROM DIM_CUENTA CU
    INNER JOIN DIM_CONTAC CO        ON CU.IDE_CUENTA=CO.IDE_CUENTA
    INNER JOIN DIM_NIVEDU_CONTAC NE ON CO.IDE_NIVEDU_CONTAC=NE.IDE_NIVEDU_CONTAC
"""
df_cliente=pd.read_sql_query(sql, conexion)
df_cliente.head(2)
```

```python
# llevamos la clave primaria al índice
df_cliente=df_cliente.set_index('IDE_CUENTA')
df_cliente.head()
```

```python
# Obtenemos dataframe con una única fila por cliente que ha comprado y los productos por columnas
df_venta_cliente_pro_num=df_venta.pivot_table(index="IDE_CUENTA",columns="DES_PRODUC",values="NUM_PRECTOT_LIN_OPORTU",fill_value=0)
df_venta_cliente_pro_num.head()
```

```python
# Obtener indicadores si consumo o no el producto
df_venta_cliente_pro_ind=df_venta_cliente_pro_num.astype(bool)
df_venta_cliente_pro_ind.head()
```

```python
# Añadimos prefijos en las dos tablas. Indicamos NUM_ (si es un consumo en €) o IND_ (si es un indicador de consumido)
df_venta_cliente_pro_num=df_venta_cliente_pro_num.add_prefix("NUM_")
df_venta_cliente_pro_ind=df_venta_cliente_pro_ind.add_prefix("IND_")
print(df_venta_cliente_pro_num.head(2))
print(df_venta_cliente_pro_ind.head(2))
```

```python
# Generamos una único dataframe con el importe de la venta y el indicador
df_venta_cliente=pd.concat([df_venta_cliente_pro_num,df_venta_cliente_pro_ind],axis=1)
df_venta_cliente.head()
```

```python
# Inner join por índice de las tabla de ventas por cuenta con la de datos por cuenta 
dp_FAC_CLIENTE_VENTA=df_cliente.merge(df_venta_cliente, left_index=True, right_index=True)
dp_FAC_CLIENTE_VENTA.head(2)
```

```python
# Guardamos esta tabla en la bbdd
dp_FAC_CLIENTE_VENTA.to_sql('ds_tr_X', conexion, index=True, if_exists='replace')
```

```python
# Identificamos la clase objetivo cambiándole el nombre y eliminamos la variable con su consumo
dp_FAC_CLIENTE_VENTA.columns=dp_FAC_CLIENTE_VENTA.columns.str.replace("^IND_PRO13","CLS_PRO13",  regex=True)
dp_FAC_CLIENTE_VENTA = dp_FAC_CLIENTE_VENTA.loc[:, ~dp_FAC_CLIENTE_VENTA.columns.str.startswith("NUM_PRO13_")]
dp_FAC_CLIENTE_VENTA.head()
```

```python
# Ponemos la clase objetivo al final
pos_cls = dp_FAC_CLIENTE_VENTA.columns.str.startswith("CLS_")
pos_res = ~pos_cls
dp_FAC_CLIENTE_VENTA=pd.concat([dp_FAC_CLIENTE_VENTA.iloc[:,pos_res], dp_FAC_CLIENTE_VENTA.iloc[:,pos_cls]], axis=1)
dp_FAC_CLIENTE_VENTA.head()
```

```python
# Guardamos los datos para modelado
dp_FAC_CLIENTE_VENTA.to_sql('ds_tr_XY', conexion, index=True, if_exists='replace')
```

```python
# Se cierra cursor bbdd
cursor.close()
```

## Preprocesado de datos para modelos de clasificación

A partir de los datos obtenidos para el modelado se generan los datos para entrenamiento y test de los diferentes modelos.

Entradas:
- **ds_tr_XY**: datos para el modelado obtenidos tras el preprocesamiento 

Salidas:
- **X_train, Y_train, X_test, Y_test**: datos para entrenamiento y test en su escala orginal (con variables tanto numéricas como categóricas). La variables objetivo o dependiente está en las tablas Y_ y las variables independientes en las X_

- **X_test_scaled**: datos para entrenamiento  en escala numérica normalizada tras aplicar _z-score_

### Obtención de entrenamiento y test en su escala original

```python
# librerías a usar
import pandas as pd
import sqlite3
import numpy as np
import pickle
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# configuración pandas
pd.set_option('display.max_columns', None)
```

```python
# apertura de la base de datos y generar cursor
conexion=sqlite3.connect("../data/final/dw_sf_tr.db")
cursor=conexion.cursor()
```

```python
# Datos resultado de la etapa de preprocesamiento
sql="""
    SELECT *
    FROM ds_tr_XY
"""
dp_FAC_CLIENTE_VENTA=pd.read_sql_query(sql, conexion, index_col="IDE_CUENTA")

# Cambios las campos DES_ a categorias y los IND_ a booleanos
col_cat = list(dp_FAC_CLIENTE_VENTA.filter(regex=r"^DES_").columns)
col_bol = list(dp_FAC_CLIENTE_VENTA.filter(regex=r"^IND_").columns)
col_cls = list(dp_FAC_CLIENTE_VENTA.filter(regex=r"^CLS_").columns)

dp_FAC_CLIENTE_VENTA[col_cat] = dp_FAC_CLIENTE_VENTA[col_cat].astype("category")
dp_FAC_CLIENTE_VENTA[col_bol] = dp_FAC_CLIENTE_VENTA[col_bol].astype("bool")
dp_FAC_CLIENTE_VENTA[col_cls] = dp_FAC_CLIENTE_VENTA[col_cls].astype("category")
dp_FAC_CLIENTE_VENTA.info()
```

```python
# Se identifica dos conjuntos de variables las indendientes (Y) y la dependiente (X)
Y=dp_FAC_CLIENTE_VENTA[col_cls]
X=dp_FAC_CLIENTE_VENTA.drop(col_cls,axis=1)
```

```python
# Se fija la semilla para que el muestreo sea reproducible
np.random.seed(123)
```

```python
# Se parte en entrenamiento y test usando muestreo estratificado por la variable objetivo
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

### Obtención datos entrenamiento en escala numérica mediante OneHotEncoding

```python
# Datos entrenamiento para clasificación en escala numérica
# aplicando la técnicas de one hot encoding

# Crear un modelo de OneHotEncoder
md_DUMMIES = OneHotEncoder(sparse=False,dtype=bool)
```

```python
# Crear un nuevo DataFrame con las variables numéricas transformando las categorías
df_encoded = md_DUMMIES.fit_transform(X_train[col_cat])
df_num = pd.DataFrame(df_encoded, columns=md_DUMMIES.get_feature_names_out(X_train[col_cat].columns),index=X_train.index)
df_num.columns = df_num.columns.str.replace(' ', '_')
```

```python
X_train_num = pd.concat([X_train.drop(col_cat,axis=1),df_num], axis=1)
X_train_num.head()
```

### Obtención de datos entrenamiento en escala numérica normalizada mediante Z-score

```python
# Modelo de preprocesamiento con Z-score obtenido desde el conjunto de entrenamiento
md_SCALE = StandardScaler()
md_SCALE.fit(X_train_num)
```

```python
# Aplicamos el módelo al conjunto de entrenamiento
df_scale = md_SCALE.transform(X_train_num)
X_train_scaled = pd.DataFrame(df_scale,columns=X_train_num.columns, index=X_train_num.index)
X_train_scaled.describe()
```

```python
# Datos test para clasificación en escala numérica
# aplicando la técnicas de one hot encoding obtenidas sobre el conjunto entrenamiento
```

```python
# Crear un nuevo DataFrame con las variables numéricas transformando las categorías
df_encoded = md_DUMMIES.fit_transform(X_test[col_cat])
df_num = pd.DataFrame(df_encoded, columns=md_DUMMIES.get_feature_names_out(X_test[col_cat].columns),index=X_test.index)
df_num.columns = df_num.columns.str.replace(' ', '_')
df_num
```

```python
# Definimos la estrategia de balanceo
balanceo = RandomUnderSampler(sampling_strategy='majority', random_state=mod_conf_random)
```

```python
# Realizar validación cruzada y cálculo de métricas
for train_index, test_index in mod_conf_cv.split(X_train, Y_train):
    
    # Balancear los datos de entrenamiento en cada iteración   
    X_train_balanced, Y_train_balanced = balanceo.fit_resample(X_train.iloc[train_index,:], Y_train[train_index])

    # Ajustar el modelo en los datos balanceados y encontrar los mejores hiperparámetros
    grid_search = GridSearchCV(mod, mod_par_grid, scoring=mod_conf_metric, cv=mod_conf_cv)
    grid_search.fit(X_train_balanced, Y_train_balanced)
    
    # Obtener el modelo con los mejores hiperparámetros
    best_mod = grid_search.best_estimator_
    
    # Predecir las clases y probabilidades en los datos sin balancear
    Y_pred = best_mod.predict(X_train.iloc[test_index, :])
    Y_prob = best_mod.predict_proba(X_train.iloc[test_index, :])[:, 1]
    
    # Calcular las métricas
    auc = roc_auc_score(Y_train[test_index], Y_prob)
    sensitivity = recall_score(Y_train[test_index], Y_pred)
    specificity = recall_score(Y_train[test_index], Y_pred, pos_label=1)
    
    # Almacenar las métricas en las listas
    mod_auc.append(auc)
    mod_sensitivity.append(sensitivity)
    mod_specificity.append(specificity)    
```

```python
# Modelo final con todos los datos balanceados
X_train_balanced, Y_train_balanced = balanceo.fit_resample(X_train, Y_train)

# se ajustan los parámetros con los datos balanceados
grid_search = GridSearchCV(mod, mod_par_grid, scoring=mod_conf_metric, cv=mod_conf_cv)
grid_search.fit(X_train_balanced, Y_train_balanced)

# Obtener el modelo con los mejores hiperparámetros y datos balanceados
best_mod = grid_search.best_estimator_
print("Mejores parametros:", grid_search.best_params_)
```

```python
# Imprimir las métricas de validacion
print("Métricas de validación:")
print(f'AUC: {np.array(mod_auc).mean()}')
print(f'Sensibilidad: {np.array(mod_sensitivity).mean()}')
print(f'Especificidad: {np.array(mod_specificity).mean()}')
```

```python
# Guardar el modelo y las predicciones en el mismo archivo pickle
mod_dict = {
    'nombre' : mod_name,
    'modelo': best_mod,
    'preprocesamiento':'zscore',     
    'auc': mod_auc,
    'sensibilidad': mod_sensitivity,
    'especificidad': mod_specificity   
}

with open('../models/'+mod_name+'.pkl', 'wb') as file:
    pickle.dump(mod_dict, file)
```
